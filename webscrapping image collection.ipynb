```python
import selenium 
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
from selenium.common.exceptions import NoSuchElementException
```

Getting the webdriver

```python
driver=webdriver.Firefox("C:\geckodriver-v0.29.1-win64")
```

Opening Amazon website

```python
driver.get("https://amazon.in")
```

# Scraping Saress Images, 

Checking for search button and entering the saress

```python
enter=driver.find_element_by_id("twotabsearchtextbox")
```

```python
enter.send_keys("Sarees")
```

Clicking on search button

```python
driver.find_element_by_id("nav-search-submit-button").click()
```

```python
#creating empty list
img_urls = []

# scrapping from at least 5 pages
for i in range(0,4):
    for m in driver.find_elements_by_xpath('//img[@class="s-image"]'):
        img_urls.append(m.get_attribute('src'))
        #moving to next pages
    try:
        driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").click()
        time.sleep(1)
    except:
        break

 #With the collected ing_urls downloading the images   
for i in range(len(img_urls)):
    if i >= 500:
        break
    print("Downloading {0} of {1} images" .format(i, 500))
    response= requests.get(img_urls[i])
    file = open(r"E:\sarees1\sarees"+str(i)+".jpg", "wb") #downloading images to particular folder
    file.write(response.content)
```
